```python
import zipfile
import pydicom
import matplotlib.pyplot as plt
from io import BytesIO

# Path to your ZIP file
zip_path = "/projects/0/aus20644/data/ukbiobank/imaging/cardiac_mri/20211_cine_tagging_images/imaging_visit_array_0/1993836_20211_2_0.zip"

# Open the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    # List all files in the ZIP
    dicom_files = [f for f in zip_ref.namelist() if f.endswith(".dcm")]
    print(len(dicom_files))

    if dicom_files:
        # Read the first DICOM file in the ZIP
        with zip_ref.open(dicom_files[59]) as dcm_file:
            ds = pydicom.dcmread(BytesIO(dcm_file.read()))

        # Display the DICOM image
        plt.imshow(ds.pixel_array, cmap="gray")
        plt.show()
    else:
        print("No DICOM files found in the ZIP.")
```
